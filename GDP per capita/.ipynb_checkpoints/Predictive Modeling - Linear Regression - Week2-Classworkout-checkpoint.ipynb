```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.formula.api import ols
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

### Linear Regression Math

```python
df = pd.DataFrame({'Weight':[2.5,2.2,3.1,3.8,4.2],'Height':[3.1,2.8,3.5,4.0,4.5]})
df
```

```python
x =df['Weight']  # IDV
y =df['Height'] # TV
```

```python
x = (x-x.mean())/x.std() # Scaling = (x-x_mean)/ x_std 
y = (y-y.mean())/y.std() 
```

```python
x
```

```python
y
```

#### StatsModel

```python
import statsmodels.api as sm
```

```python
x_cons = sm.add_constant(x) # Manually we add the constant. So that i get the intercept in the model.
x_cons
```

```python
lr_ols = sm.OLS(y,x_cons).fit()  # GD will be implemented and get the best bo and b1
lr_ols.summary()
```

#### Manual Gradient Descent

```python
# Gradient Descent

# S1 : Radomly initialize weight
b0 = 0
b1 = 0
epochs =1000
n = len(df)
y_act = y
alpha=0.1
b0_list=[]
b1_list=[]
sse_list=[]
for i in range(epochs): # S5: Repeat S2,S3,S4
    b1_list.append(b1)
    b0_list.append(b0)
 # S2 : Calculation y_pred   
    y_pred = (b1 * x)+ b0
# S3 : Loss calulation
    sse = np.power(np.sum(y_act-y_pred),2)
    sse_list.append(sse)
    der_b1 = (-1/n)* np.sum(x*(y_act-y_pred))
    der_b0 = (-1/n)* np.sum((y_act-y_pred))
# S4 : Updating the beta values    
    b1 = b1 -(alpha * der_b1)
    b0= b0 -(alpha*der_b0)
    
print('Slope:',b1)
print('Intercept:',b0)
```

```python
plt.scatter(b1_list,sse_list)
plt.show()
```

```python
# Gradient Descent

# S1 : Radomly initialize weight
b0 = 0
b1 = 10
epochs =1000
n = len(df)
y_act = y
alpha=0.1
b0_list=[]
b1_list=[]
sse_list=[]
for i in range(epochs): # S5: Repeat S2,S3,S4
    b1_list.append(b1)
    b0_list.append(b0)
 # S2 : Calculation y_pred   
    y_pred = (b1 * x)+ b0
# S3 : Loss calulation
    sse = np.power(np.sum(y_act-y_pred),2)
    sse_list.append(sse)
    der_b1 = (-1/n)* np.sum(x*(y_act-y_pred))
    der_b0 = (-1/n)* np.sum((y_act-y_pred))
# S4 : Updating the beta values    
    b1 = b1 -(alpha * der_b1)
    b0= b0 -(alpha*der_b0)
    
print('Slope:',b1)
print('Intercept:',b0)
```

```python
plt.scatter(b1_list,sse_list)
plt.show()
```

```python
for column in df.columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3-q1
    ll = q1 - (1.5*iqr)
    ul = q3 + (1.5*iqr)
    for index in df[column].index:
        if df.loc[index,column]> ul:
            df.loc[index,column]=ul
        if df.loc[index,column]< ll:
            df.loc[index,column]=ll
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
# Training - 80%, Testing = 20% (Random selection of train_test_split)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=10)
```

## Linear Regression Model

```python
x_train_cons = sm.add_constant(x_train)
x_train_cons.head()
```

```python
# Adding constant will give me the intercept of the model
```

```python
lin_reg_ols = sm.OLS(y_train,x_train_cons).fit()
lin_reg_ols.summary()
```

## Performance Metrics

```python
x_test_cons = sm.add_constant(x_test)
y_test_pred = lin_reg_ols.predict(sm.add_constant(x_test_cons)) # Predicted values of x_test
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error
```

```python
mse = mean_squared_error(y_test,y_test_pred) # actual and predicted
print('MSE:',mse)
print('RMSE:',np.sqrt(mse))
mape = mean_absolute_percentage_error(y_test,y_test_pred)
print("MAPE:",mape*100)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_test_pred)  # Test R2
```

## Train & Test Score

```python
print('Train RMSE:',np.sqrt(mean_squared_error(y_train,y_train_pred)))
print('Test RMSE:',np.sqrt(mean_squared_error(y_test,y_test_pred)))
```

```python
print('Train MAPE:',np.sqrt(mean_absolute_percentage_error(y_train,y_train_pred)))
print('Test MAPE:',np.sqrt(mean_absolute_percentage_error(y_test,y_test_pred)))
```

```python
print('Train R2:',r2_score(y_train,y_train_pred))
print('Test R2:',r2_score(y_test,y_test_pred))
```
