### Separate columns having (-)ve values

```python
columns_with_negatives=[]
for col in df.columns:
    if df[col].dtype != object :
        if (df[col] < 0).any():
            columns_with_negatives.append(col)
print(columns_with_negatives)
```

### "As per the above output, the columns with negative values are 'Time spent playing outdoor sports', 'Total Time spent working in front of the screen', 'Sleeping hours', and 'Number of friends.' For these columns, negative values are not possible; they can be zero but not negative."
#

### Negative values count 

```python
for col in columns_with_negatives:
    print(col," : ")
    count=0
    for index in df[col].index:            # We can also use "df.index" instead of "df[col].index" for both indices are the same.
        if df.loc[index, col] < 0 :
            count += 1
    print(count)
    
```

### The size of the incorrect data or data with negative values is relatively small. It could be due to printing mistakes or other issues, and I don't have enough information about the data source to determine the cause. Therefore, I have two options:

### 1. Remove this data: Given the limited understanding of the nature of the negative values and their potential impact on the analysis, I lean towards removing this data from the dataset to ensure data integrity and avoid introducing potential inaccuracies.

### 2.Replace this data: While there are various methods for data replacement, my lack of knowledge about the data and its sources makes me cautious about making replacements. 

### As a result, I'm inclined to remove the data to maintain data quality and reliability.

## Outlier Analysis for Continuous Data

```python
for col in Cont_columns:
    sns.boxplot(x = col, data = df)
    plt.show()
    
    # Identify Quartile 
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define a threshold for outliers (you can adjust this threshold)
    threshold = 1.5
    
    # Identify lower limit and upper limit
    ll = Q1 - IQR*threshold
    ul = Q3 + IQR*threshold
    
    # Identify outliers
    outlier = df[(df[col] < ll) | (df[col] > ul)]
    
    print("Quartile 1  : ",Q1)
    print("Quartile 3  : ",Q3)
    print("Lower_limit : ",ll)
    print("Upper limit : ",ul)
    print("Outlier     : ",len(outlier))
    
    print("\n","*"*75,"\n")
    
```
